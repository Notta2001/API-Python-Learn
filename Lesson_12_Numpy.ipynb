# Copy data

```python
a_list = [1, 3, 5]

list1 = a_list
list2 = a_list[:]
list3 = a_list.copy()
```

```python
a_list[0] = 9
```

```python
list1, list2, list3
```

List1 thay đổi vì nó trỏ cùng một vị trí

List2 khi slicing thì copy ra luôn chứ k view như numpy

List3 là copy hoàn toàn

```python
data = np.array([1, 2, 3])

data1 = data
data2 = data[:]
data3 = data.copy()
```

```python
data[0] = 9
```

```python
data1, data2, data3
```

Data1 và data2 chỉ thay đổi góc nhìn nên thay đổi

Data3 copy hoàn toàn

# One hot vector

```python
import numpy as np

num_labels = 10
num_samples = 12
data = np.random.randint(0, num_labels, num_samples)
data
```

```python
labels_oh = np.zeros((num_samples, num_labels))
labels_oh
```

```python
first_indices = np.arange(num_samples)
labels_oh[first_indices, data]  =1
labels_oh
```
